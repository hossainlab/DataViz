# How do you speak Matplotlib?

A *matplotlib* visualization is a `figure` onto which is attached one or more `axes`. Each `axes` has a horizontal (x) `axis` and vertical (y) `axis`, and the data is encoded using color and glyphs such as `markers` (for example circles) or `lines` or polygons (called `patches`). The figure below annotates these parts of a visualization and was created by Nicolas P. Rougier using `matplotlib`. The source code can be found in the [matplotlib documentation](https://matplotlib.org/gallery/showcase/anatomy.html#sphx-glr-gallery-showcase-anatomy-py).

# Import Library by Convention

```python
import numpy as np # for linear algebra 
import matplotlib.pyplot as plt # for plotting 
import pandas as pd # for data analysis 
```

# Line Plot 

```python
# Create data 
x = np.linspace(0, 2*np.pi, 50)
y = np.sin(x)
```

## Let's make our first line plot

```python
plt.plot(y)
plt.show() 
```

## Let's customize plot 

```python
# Colors, marker and line style 
plt.plot(y, color = 'green', marker='o', linestyle='-') 
plt.show() 
```

```python
# We can define same thing above inside a single string 
plt.plot(y, 'go-') # g = greem, o = circle, - = line style 
plt.show() 
```

```python
# Markersize and line width 
plt.plot(y, 'go-', markersize=8, linewidth=5)
plt.show() 
```

```python
# Let's make a title, xlabel, ylabel of our plot 
plt.plot(y, 'go-', markersize=8, linewidth=5)
plt.title("$sin(x)$") # $sin(x)$ LaTeX formatting
plt.xlabel("Radians")
plt.ylabel("Amplitude")
plt.show() 
```

```python
# Let's change the fontsize of title, xlabel and ylabel 
# fontsize = 'large' or 'medium' or 'small'
plt.plot(y, 'go-', markersize=8, linewidth=5)
plt.title("$sin(x)$", fontsize='large') 
plt.xlabel("Radians", fontsize='large')
plt.ylabel("Amplitude", fontsize='large')
plt.show() 
```

```python
# We can change the fontsize of title, xlabel and ylabel 
plt.plot(y, 'go-', markersize=8, linewidth=5)
plt.title("$sin(x)$", fontsize=15) 
plt.xlabel("Radians", fontsize=12)
plt.ylabel("Amplitude", fontsize=12)
plt.show() 
```

## Let's make a multiple line plot in a figure

```python
# Create a vector called y  
y = np.linspace(0, 10, 20)
y1 = x**2.0
y2 = x**1.5
plt.plot(y, y1, 'bo-', linewidth=2, markersize=8) # 'bo-' = blue, circle, solid 
plt.plot(y, y2, 'gs-', linewidth=2, markersize=8) # 'gs-' = green, square, solid 
plt.title("Plot of $X$ and $Y$", fontsize=12)
plt.xlabel("$X$", fontsize=12)
plt.ylabel("$Y$", fontsize=12)
plt.show() 
```

```python
# Add label and legend 
y = np.linspace(0, 10, 20)
y1 = x**2.0
y2 = x**1.5
plt.plot(y, y1, 'bo-', linewidth=2, markersize=8, label="First") # 'bo-' = blue, circle, solid 
plt.plot(y, y2, 'gs-', linewidth=2, markersize=8, label="Second") # 'gs-' = green, square, solid 
plt.title("Plot of $X$ and $Y$", fontsize=12)
plt.xlabel("$X$", fontsize=12)
plt.ylabel("$Y$", fontsize=12)
plt.legend(loc="upper left") # Location of legend 
plt.show() 
```

```python
# Add Padding 
y = np.linspace(0, 10, 20)
y1 = x**2.0
y2 = x**1.5
plt.plot(y, y1, 'bo-', linewidth=2, markersize=8, label="First") # 'bo-' = blue, circle, solid 
plt.plot(y, y2, 'gs-', linewidth=2, markersize=8, label="Second") # 'gs-' = green, square, solid 
plt.title("Plot of $X$ and $Y$", fontsize=12)
plt.xlabel("$X$", fontsize=12)
plt.ylabel("$Y$", fontsize=12)
plt.legend(loc="upper left") # Location of legend 
plt.tight_layout() 
plt.show() 
```

```python
# List available matplotlib style 
print(plt.style.available)  
```

```python
# Add style  
y = np.linspace(0, 10, 20)
y1 = x**2.0
y2 = x**1.5

plt.style.use('ggplot')
plt.plot(y, y1, 'bo-', linewidth=2, markersize=8, label="First") # 'bo-' = blue, circle, solid 
plt.plot(y, y2, 'gs-', linewidth=2, markersize=8, label="Second") # 'gs-' = green, square, solid 
plt.title("Plot of $X$ and $Y$", fontsize=12)
plt.xlabel("$X$", fontsize=12)
plt.ylabel("$Y$", fontsize=12)
plt.legend(loc="upper left") # Location of legend 
plt.tight_layout() 
plt.show() 
```

```python
# Add style  
y = np.linspace(0, 10, 20)
y1 = x**2.0
y2 = x**1.5

plt.style.use('seaborn')
plt.plot(y, y1, 'bo-', linewidth=2, markersize=8, label="First") # 'bo-' = blue, circle, solid 
plt.plot(y, y2, 'gs-', linewidth=2, markersize=8, label="Second") # 'gs-' = green, square, solid 
plt.title("Plot of $X$ and $Y$", fontsize=12)
plt.xlabel("$X$", fontsize=12)
plt.ylabel("$Y$", fontsize=12)
plt.legend(loc="upper left") # Location of legend 
plt.tight_layout() 
plt.show() 
```

```python
# Add style  
y = np.linspace(0, 10, 20)
y1 = x**2.0
y2 = x**1.5

plt.style.use('fivethirtyeight')
plt.plot(y, y1, 'bo-', linewidth=2, markersize=8, label="First") # 'bo-' = blue, circle, solid 
plt.plot(y, y2, 'gs-', linewidth=2, markersize=8, label="Second") # 'gs-' = green, square, solid 
plt.title("Plot of $X$ and $Y$", fontsize=12)
plt.xlabel("$X$", fontsize=12)
plt.ylabel("$Y$", fontsize=12)
plt.legend(loc="upper left") # Location of legend 
plt.tight_layout() 
plt.show() 
```

```python
# Save Plot 
y = np.linspace(0, 10, 20)
y1 = x**2.0
y2 = x**1.5

plt.style.use('ggplot')
plt.plot(y, y1, 'bo-', linewidth=2, markersize=8, label="First") # 'bo-' = blue, circle, solid 
plt.plot(y, y2, 'gs-', linewidth=2, markersize=8, label="Second") # 'gs-' = green, square, solid 
plt.title("Plot of $X$ and $Y$", fontsize=12)
plt.xlabel("$X$", fontsize=12)
plt.ylabel("$Y$", fontsize=12)
plt.legend(loc="upper left") # Location of legend 
plt.tight_layout() 
plt.savefig("plot.pdf") # .png, .jpg, .tiff, .jepg; whatever you want
plt.show() 
```
