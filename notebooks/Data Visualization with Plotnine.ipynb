```python
import pandas as pd 
import numpy as np 
from plotnine import *
from plotnine.data import * 
import seaborn as sns 
iris = sns.load_dataset('iris')
%matplotlib inline 
```

```python
iris.head() 
```

```python
(ggplot(iris) 
 +geom_point(aes(x='sepal_length', y='sepal_width', fill='species'))
)
```

```python
(ggplot(iris) 
 +geom_boxplot(aes(x='species', y='sepal_width'))
)
```

```python
(ggplot(iris) 
 +geom_boxplot(aes(x='species', y='petal_width'))
)
```
