# Data Visualization with Seaborn and Matplotlib

```python
# We first need to import the packages that we will be using
import seaborn as sns # For plotting
import matplotlib.pyplot as plt # For showing plots
# plt.style.use('ggplot')
sns.set_style('darkgrid')
# Load in the data set
titanic_data = sns.load_dataset("titanic")

%matplotlib inline 
```

## Data Visualization


```python
pal = dict(man="#003f5c", woman="#2f4b7c", child="#665191", male="#a05195", female="#d45087", 
           adult_male="f95d6a")
```

```python
# Who were titanic passengers? How many men, women and children are in column?(sex column)
sns.catplot('sex', kind='count', data=titanic_data, palette=pal)
plt.title("No. of Male and Female")
plt.xlabel("Sex")
plt.ylabel("Frequency")
plt.show()
```

```python
# Who were titanic passengers? How many men, women and children are in column?(who column)
sns.catplot('who', kind='count', data=titanic_data)
plt.title("No. of Men, Women and Children")
plt.xlabel("Who")
plt.ylabel("Frequency")
plt.show()
```

```python
# Who were titanic passengers? How many men, women and children are in column?
# Separate by passenger class 
sns.catplot('sex', kind='count', hue='class', data=titanic_data)
plt.title("No. of Male and Female Depending on Class")
plt.xlabel("Sex")
plt.ylabel("Frequency")
plt.show()
```

```python
# Who were titanic passengers? How many men, women and children are in column?(who column)
# Separate by passenger class
sns.catplot('who', kind='count',hue='class', data=titanic_data)
plt.title("No. of Men, Women and Children Depending on Class")
plt.xlabel("Who")
plt.ylabel("Frequency")
plt.show()
```

```python
# Who were titanic passengers? How many adults are in column?
sns.catplot('adult_male', kind='count',data=titanic_data)
plt.title("No. of Adult Male")
plt.xlabel("Adult Male")
plt.ylabel("Frequency")
plt.show()
```

```python
# Who were titanic passengers? How many adults are in column?
# Separate by passenger class
sns.catplot('adult_male', kind='count',hue='class', data=titanic_data)
plt.title("No. of Adult Male")
plt.xlabel("Adult Male")
plt.ylabel("Frequency")
plt.show()
```

##  Distribution of Ages 

```python
# Next let's look at the distribution of ages within the groups we defined above
g = sns.FacetGrid(titanic_data, hue='sex', aspect=3)
g = g.map(sns.kdeplot, 'age', shade=True).set(xlim=(0, 80)).add_legend();
```
