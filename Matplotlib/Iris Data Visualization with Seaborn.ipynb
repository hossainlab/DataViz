# Data Visualization with Seaborn

```python
# We first need to import the packages that we will be using
import seaborn as sns # For plotting
import matplotlib.pyplot as plt # For showing plots
# plt.style.use('ggplot')
sns.set_style('darkgrid')
# Load in the data set
iris_data = sns.load_dataset("iris")

%matplotlib inline 
```

## Univariate Analysis

### Histograms

```python
# Plot a histogram of the sepal length
sns.distplot(iris_data['sepal_length'], kde=False).set_title("Distribution of Sepal Length"); 
```

```python
# Plot a histogram of the petal length
sns.distplot(iris_data['petal_length'], kde=False).set_title("Distribution of Petal Length"); 
```

```python
# Plot a histogram of the petal length and sepal length
sns.distplot(iris_data['sepal_length'], kde=False)
sns.distplot(iris_data['petal_length'], kde=False).set_title("Distribution of Sepal and Petal Length"); 
```

```python
# Plot a histogram of the sepal width
sns.distplot(iris_data['sepal_width'], kde=False).set_title("Distribution of Sepal Width"); 
```

```python
# Plot a histogram of the petal width 
sns.distplot(iris_data['petal_width'], kde=False).set_title("Distribution of Petal Width"); 
```

```python
# Plot a histogram of the sepal width and petal width
sns.distplot(iris_data['sepal_width'], kde=False)
sns.distplot(iris_data['petal_width'], kde=False).set_title("Distribution of Sepal and Petal Width"); 
```

### Grouped Plot

```python
# Create a boxplot and histogram of the petal length grouped by species
sns.boxplot(x = iris_data["species"], y = iris_data["petal_length"])

g = sns.FacetGrid(iris_data, row = "species")
g = g.map(plt.hist, "petal_length")
plt.show()
```

```python
# Create a boxplot and histogram of the petal width grouped by species
sns.boxplot(x = iris_data["species"], y = iris_data["petal_width"])

g = sns.FacetGrid(iris_data, row = "species")
g = g.map(plt.hist, "petal_width")
plt.show()
```

```python
# Create a boxplot and histogram of the sepal length grouped by species
sns.boxplot(x = iris_data["species"], y = iris_data["sepal_length"])

g = sns.FacetGrid(iris_data, row = "species")
g = g.map(plt.hist, "sepal_length")
plt.show()
```

```python
# Create a boxplot and histogram of the sepal width grouped by species
sns.boxplot(x = iris_data["species"], y = iris_data["sepal_width"])

g = sns.FacetGrid(iris_data, row = "species")
g = g.map(plt.hist, "sepal_width")
plt.show()
```

## Bivariate Analysis

### Scatter Plots

```python
# Relationship of petal width and petal length in different species
sns.scatterplot(x=iris_data['petal_width'], y=iris_data['petal_length'], hue=iris_data['species']);
```

```python
# Relationship of sepal width and petal width in different species
sns.scatterplot(x=iris_data['sepal_width'], y=iris_data['sepal_length'], hue=iris_data['species']);
```

### Pairplots

```python
# Pairplot of iris dataset
sns.pairplot(iris_data);
```

```python
# Pairplot of iris dataset depending on species 
sns.pairplot(iris_data, hue="species");
```

```python
# Pairplot of iris dataset using color palette `husl`
sns.pairplot(iris_data, hue="species", palette="husl");
```

```python
# Pairplot of iris dataset using markers
sns.pairplot(iris_data, hue="species", markers=["o", "s", "D"]);
```

```python
# Pairplot of iris dataset depending on variables
sns.pairplot(iris_data, vars=["sepal_width", "sepal_length"]);
```

```python
# Larger pairplot
sns.pairplot(iris_data, height=3, vars=["sepal_width", "sepal_length"]);
```

```python
# Pairplot of iris dataset by using x-variables and y-variables(passing hue)
sns.pairplot(iris_data, size=4, hue='species',
             x_vars=["sepal_width", "sepal_length"],y_vars=["petal_width", "petal_length"]); 
```

```python
# Pairplot of iris dataset by using x-variables and y-variables
sns.pairplot(iris_data, size=4, x_vars=["sepal_width", "sepal_length"],y_vars=["petal_width", "petal_length"]); 
```

```python
# Density pairplot
sns.pairplot(iris_data, diag_kind='kde');
```

```python
# Linear regression models to scatter plot
sns.pairplot(iris_data, kind='reg');
```

```python
# More functionality on pairplot
sns.pairplot(iris_data, diag_kind="kde", markers="+",size=4,
                 plot_kws=dict(s=50, edgecolor="b", linewidth=1),
                 diag_kws=dict(shade=True));
```
