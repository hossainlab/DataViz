# Introduction to Seaborn
- **Seaborn** is a statistical plotting library.
- It has beautiful default styles. 
- It is also designed to work very well with pandas **DataFrame** objects. 

__Documentation:__
[Official seaborn tutorial](https://seaborn.pydata.org/tutorial.html#tutorial)

```python
# imports 
import matplotlib.pyplot as plt 
import numpy as np 
import seaborn as sns 
```

```python
# load seaborn built-in dataset
tips = sns.load_dataset('tips')
```

## Exploring Dataset

```python
# examine first few rows 
tips.head() 
```

```python
# summary 
tips.info() 
```

## Distribution Plot: Univariate Analysis

```python
# KDE plot default
sns.distplot(tips['total_bill'])
```

```python
# remove KDE 
sns.distplot(tips['total_bill'], kde=False)
```

```python
# change no. of bins 
sns.distplot(tips['total_bill'], kde=False, bins=30)
```

## Joint Plots: Bivariate Analysis 

```python
# joint plot 
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
# joint plot: kind 
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
# joint plot: kind 
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
# joint plot: kind 
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

## Pair Plot

```python
sns.pairplot(tips)
```

```python
# hue: categorical variable  
sns.pairplot(tips, hue='sex')
```

```python
# color palette
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

## Rugplot

```python
sns.rugplot(tips['total_bill'])
```

```python
# compare with distplot 
sns.distplot(tips['total_bill'], kde=False)
```

```python
# kde plot 
sns.kdeplot(tips['total_bill'])
```

## Categorical Plots

```python
sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
# estimator 
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.median)
```

```python
# estimator 
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

## Count Plot 

```python
sns.countplot(x='sex', data=tips)
```

## Box Plots 

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
# comparison: hue = categorical 
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

## Violin Plot

```python
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
# hue 
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
# split 
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```
